```python
%matplotlib inline
import re
import warnings
import math
from datetime import timedelta

import matplotlib as mpl
import matplotlib.animation as animation
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
import squarify
```

- **BS**: *(float)* Bit size
- **GR**: *(float)* Gamma Ray
- **NEUT**: *(float)* Neutron
- **DENS**: *(float)* Density
- **SP**: *(float)* Spontaneous Potential
- **PEF**: *(float)* Photoelectric Effect
- **DTC**: *(float)* Compressional Slowness
- **DRHO**: *(float)* Density-Porosity
- **RESD**: *(float)* Resistivity Deep
- **RESS**: *(float)* Resistivity Shallow
- **RESM**: *(float)* Resistivity Medium
- **CALI**: *(float)* Caliper
- **TENS**: *(float)* Tension
- **TEMP**: *(float)* Temperature
- **DEPT**: *(float)* Depth of the data point measured along the well
- **ONSHORE**: *(bool)* Boolean value indicating if the data point belongs to an onshore well (if not, it belongs to an offshore well)
- **DIRSURVEY**: *(bool)* Boolean value indicating if there was any directional survey available to infer the correct position of this data point
- **WELLNAME**: *(str)* Name of the well containing the given data point

```python
df = pd.read_csv('C:/Users/sotla/logs.csv', dtype={'WELLNAME': str}, low_memory=False)


columns_to_plot = ['BS', 'GR', 'DENS', 'SP', 'TEMP', 'PEF', 'DEPT']

plt.figure(figsize=(15, 10))
for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(2, 4, i)
    plt.hist(df[column].dropna(), bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Гистограмма для {column}')
    plt.xlabel(column)
    plt.ylabel('Частота')

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('C:/Users/sotla/logs.csv', dtype={'WELLNAME': str}, low_memory=False)

parameters_to_plot = ['GR', 'DENS', 'SP', 'PEF', 'TEMP', 'RESD', 'RESS']
plt.figure(figsize=(15, 5 * len(parameters_to_plot)))
for i, param in enumerate(parameters_to_plot, 1):
    plt.subplot(len(parameters_to_plot), 1, i)
    sns.boxplot(data=df, x='WELLNAME', y=param)
    plt.title(f'Распределение {param} по скважинам')
    plt.xlabel('Скважина')
    plt.ylabel(param)
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
